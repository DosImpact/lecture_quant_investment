#### `reindex()`의 유용한 Arguments

- `fill_value`

```python
s2 = s.copy() # 깊은 복사
s2
```

```python
s2.reindex(['a', 'f']) # a,f 만 가져온다.
```

```python
s2.reindex(['a', 'f'], fill_value=0)  # fill 0 insteand of Nan
```

- `method`

```python
s3 = pd.Series(['red', 'green', 'blue'], index=[0, 3, 5])
s3
```

```python
s3.reindex(np.arange(0,7)) # 0~7의 인덱스로 확장
# 참고로 nan은 숫자 연산도중 발생한 상황
# NaN은 dtype이 object에서 발생한 상황황
```

```python
# NaN을 발생시키지 말고 이전 값으로 채워라# NaN을 발생시키지 말고 이전 값으로 채워라
s3.reindex(np.arange(0,7), method='ffill') # forward fill - nan을 이전값으로 매꿔라.
s3.reindex(np.arange(0,7), method='ffill') # forward fill - nan을 이전값으로 매꿔라.
```
