## 연산 기준

- DataFrame은 기준이 columns
- Series는 기준이 index
- 따로 명시가 없다면 Series의 index가 DataFrame의 columns에 맞춰짐!

## DataFrame & Series

```python
# 첫번째 row로 모든 row를 빼보자 ( 노말라이즈 )

price_df.iloc[0] # 첫번째 데이터를 가져왔다.
# 해당 row로 나머지 row를 뺄건데, normalize 느낌으로 진행하고싶다.
```

```python
# Subtract row Series
# DataFrame의 기준인 columns와 Series의 기준인 index가 서로 일치하기 때문에, 의도한 대로 계산 가능
(price_df - price_df.iloc[0]).head() # df타입에 series가 브로드 케스팅 되어서, 각 row가 뺄셈이된다.
```

```python
price_df['Open'].head()
```

```python
# Subtract column Series
# [X] DataFrame의 기준인 columns와 Series의 기준인 index가 서로 불일치 (price_df['open'] - price_df도 마찬가지로 [X] )
(price_df - price_df['Open']).head() # 카테시안 프로덕트로 연산되고, nan이 나오게 된다. ( 데이터 얼라인이 안맞아 의도한 결과,아님)
```

## DataFrame & DataFrame

- index,column 이 일치하는 것 끼리만 element-wise 연산이 이루어지고 나머지는 nan 처리
- element-wise : 얼라인이 되면 연산하고, 아니라면 nan 임

```python
price_df - price_df[['Open', 'Low']].iloc[:2] 
# open,low 중 2번째까지만 element-wise 된것을 볼 수 있다. 
```

## 연산 관련 pandas built-in 함수

- axis란?
- 연산은 기본적으로 "axis를 변형(줄이거나 늘리는)하는 방식" 으로 진행된다.
* axis=0 으로 연산한 결과 axis=0이 없어진다.
- 1차 접근, axis == 0, 아래로 밀어버려 , x가 증가하는 방향으로 연산한다.

```python
#### numpy로 맛보기

import numpy as np

a = np.array([1,2,3])
b = np.array([1,2,3])
a
b
```

```python
a + b # 기본 덧셈은 axis = 0 이다.
```

```python
np.sum(
    [a,b],
    axis=0
)
```

```python
np.sum(
    [a,b],
    axis=1
)
```

```python
"""
 axis=0 --->       0            1
 axis=1 --->    0  1  2      0  1  2
            [  [1, 2, 2],   [1, 2, 2] ]
"""

"""
axis=0 방향으로 더한다는 의미
= axis=0을 없애고 싶다.변형(덧셈)해서 한 차원 줄이게 된다.(2차원>1차원 차원축소)
= axis 들이 한칸씩 밀리게 된다.

*axis=0 --->    0  1  2   
               [2, 4, 4],   
"""

"""
*axis=0 --->    0    1
             [  5,   5 ]          
"""
a = np.array([[1,2,3],[2,3,4]])
a.sum(axis=0) # array([3, 5, 7])
a.sum(axis=1) # array([6, 9])
```

```python
"""
 axis=0 --->              0                        1
 axis=1 --->      0       1     2          0       1      2
 axis=2 --->     0  1    0, 1    0        0, 1    0, 1    0, 1
            [  [[1, 2], [2, 4], [2]],   [[1, 2], [2, 2], [2, 2]] ]
"""
```

### Example
- nan은 처내고, intersection 할래

```python
_df = pd.DataFrame({'a':[1,np.nan,3], 'b':[np.nan, 2, 3]})
_df.head()
```

- 둘다 nan이 아닌 값들만 추출

```python
_df['a'].notnull() # nan아닌 알맹쓰만
_df['b'].notnull()
```

```python
# 1. a에서 nan이 아닌것이 True & b에서도 nan이 아닌것이 True
_df[ _df['a'].notnull() & _df['b'].notnull() ]
```

```python
# 2. 위 연산을 all로 가능, axis=1로 주고 all 때리면 
# true로 살아남은것이 a,b둘다 nan이 아닌것
```

```python
_df.notnull().all(axis=1)
```

```python
_df[_df.notnull().all(axis=1)] 
```

```python
# 3. dropna - 하나라도 nan이면 쳐내기 (데이터가 아깝네)
_df.dropna()
```

```python
# subset에 있는 컬럼 중에 하나라도(혹은 전부, arg로 선택가능) null이면 drop한다
_df.dropna(subset=['a'])
# 하지만 b에는 nan 이 있어도 살려주자. (a열에 nan이 있으면 쳐낸다.)
```
