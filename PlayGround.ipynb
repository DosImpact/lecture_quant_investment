### DF 요약 - 1 eg

```
df1 = pd.DataFrame({
    'c1':[1,2,3],
    'c2':[3,4,5],
    'c3':[7,8,9]
},index=range(0,3,1))
df2 = pd.DataFrame([[1,2,3],[4,5,6],[7,8,9]],
                   index=range(0,3,1),
                   columns=['c1','c2','c3']
                  )
df1.index,df2.index # RangeIndex 타입으로
df3 = df1+df2 # 인덱스 얼라인 되어 연산 가능 
df3.describe() # 개요
df3.index = df3.index.astype(str) # index 형변환
df3 = df3.set_index('c1') # c1 컬럼을 인덱스로 임시지정
df3 = df3.reset_index() # 인덱스 리셋 (초기인덱스는 제거됨)
```
